## Подготовка данных

### Названия столбцов

**Исходные названия столбцов и их значения:**
1. Name — название игры
2. Platform — платформа
3. Year_of_Release — год выпуска
4. Genre — жанр игры
5. NA_sales — продажи в Северной Америке (миллионы проданных копий)
6. EU_sales — продажи в Европе (миллионы проданных копий)
7. JP_sales — продажи в Японии (миллионы проданных копий)
8. Other_sales — продажи в других странах (миллионы проданных копий)
9. Critic_Score — оценка критиков (максимум 100)
10. User_Score — оценка пользователей (максимум 10)
11. Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Названия столбцов приводим к нижнему регистру.

```python
# исходные названия столбцов датафрейма
columns_initial = data.columns
#print(columns_initial)

#приведение названий к нижнему регистру
columns_final = columns_initial.str.lower()
#print(columns_final)

# создаем словарь для замены названий столбцов
columns_to_rename = dict(zip(columns_initial,columns_final))
#print(columns_to_rename)

#выполняем замену названий столбцов и проверяем результат
data= data.rename(columns = columns_to_rename)
data.head(3)
```

### Преобразование  типов данных

**Год выпуска**

```python
# меняем тип данных на int  в связи с целочисленным характером данных в этом столбце.
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

**Оценка критиков по 100-балльной шкале (critic_score)**

```python
# меняем тип данных на int  в связи с целочисленным характером данных в этом столбце.
data['critic_score'] = data['critic_score'].astype('Int64')
#data.info()
#data['critic_score'].unique() 
```

**Оценка пользователей по 10-балльной шкале (user_score)**

```python
# выведем все уникальные значения столбца с оценками пользователей
data['user_score'].unique()
```

Оценка 'tbd' - 'To Be Determined', то есть "Будет определено". Для данных игр на текущий момент пользовательский рейтинг еще не сформирован.

Количество данных с оценкой 'tbd'

```python
print(data[data['user_score']=='tbd']['user_score'].count())
```

Распределение по годам выпуска игр с ожидаемой пользовательской оценкой 

```python
print(data[data['user_score']=='tbd'].groupby('year_of_release')['year_of_release'].count())
```

Игры с оценкой 'tbd' представлены не только в последних годах выпуска, но и за пару десятилетий до текущего момента. Возможно, что для формирования рейтинга не хватает количества оценок. 

С точки зрения текущего момента, 'tbd' - такой же пропуск в данных о рейтингах, как и прочие пропуски, так как значение рейтинга неизвестно.

Для сохранении информации для будущих исследований,  информацию о маркере 'tbd' вынесем в отдельный столбец.  Для приведения в порядок данных, из столбца с пользовательской оценкой уберем эти данные и приведем его типу float.


```python
# выносим информацию о маркере 'tbd'  в отдельный столбец
data['tbd'] = data['user_score']=='tbd'

#меняем строковый тип данных на float с заменой некорректных значений в NaN
data['user_score'] = pd.to_numeric(data['user_score'],errors = 'coerce')
```

Рейтинги ESRB, примененные в датасете:
1. E («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым.
2. T («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. 
3. M («Mature») — «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет. 
4. E10+ («Everyone 10 and older») — «Для всех от 10 лет и старше».
5. K-A ("Kids to Adults")- первоначальное (до 1997) название текущего рейтинга E.
6. AO («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет.
7. EC («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. 
8. RP («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу

Так как не все игры могут быть с рейтингами ESRB (как частный случай,например, - RP («Rating Pending») — «Рейтинг ожидается»), а применть синтезированное по какой-либо другой категории значение может быть некорректным и внесет искажения в датасет, то заменим пропуски меткой NORATING.

 Введем ограничение для количества продаж и оценим количество аномальных значений. Определим лимиты как Q3 + 1.5*IQR

```python
data_describe = pd.DataFrame(data.describe())
data_describe = data_describe.reset_index()
data_describe = data_describe.rename(columns={'index':'statistics'})
data_describe
```

```python
# определим лимиты как Q3 + 1.5*IQR
na_limit = data_describe.loc[6,'na_sales'] + 1.5*(data_describe.loc[6,'na_sales'] - data_describe.loc[4,'na_sales'])
eu_limit = data_describe.loc[6,'eu_sales'] + 1.5*(data_describe.loc[6,'eu_sales'] - data_describe.loc[4,'eu_sales'])
jp_limit = data_describe.loc[6,'jp_sales'] + 1.5*(data_describe.loc[6,'jp_sales'] - data_describe.loc[4,'jp_sales'])
other_limit = data_describe.loc[6,'other_sales'] + 1.5*(data_describe.loc[6,'other_sales'] -data_describe.loc[4,'other_sales'])
print("Лимиты:")
print("na_limit:", na_limit)
print("eu_limit:", eu_limit)
print("jp_limit:", jp_limit)
print("other_limit:", other_limit)
```

```python
# считаем количество аномальных данных по продажам и опеределяем долю аномалий в общем к датасету
na_limit_count = data[data['na_sales']> na_limit]['na_sales'].count()
eu_limit_count = data[data['eu_sales']> eu_limit]['eu_sales'].count()
jp_limit_count = data[data['jp_sales']> jp_limit]['jp_sales'].count()
other_limit_count = data[data['other_sales']> other_limit]['other_sales'].count() 

ratio_limit = round( 100*(na_limit_count+eu_limit_count+jp_limit_count + other_limit_count)/len(data),2)
print("Отношение количества данных с  аномалиями ко всему датасету:  ",ratio_limit, "%")
```

```python
# считаем сумму продаж аномальных данных  и опеределяем долю  в общем к датасету
na_limit_sum = data[data['na_sales']> na_limit]['na_sales'].sum()
eu_limit_sum = data[data['eu_sales']> eu_limit]['eu_sales'].sum()
jp_limit_sum = data[data['jp_sales']> jp_limit]['jp_sales'].sum()
other_limit_sum = data[data['other_sales']> other_limit]['other_sales'].sum() 
all_sum = data['na_sales'].sum() + data['eu_sales'].sum() + data['jp_sales'].sum() + data['other_sales'].sum() 
#print(all_sum)
ratio_limit_sum = round( 100*(na_limit_sum + eu_limit_sum + jp_limit_sum + other_limit_sum)/all_sum,2)
print("Отношение суммы аномальных продаж  к продажам по всему датасету:  ",ratio_limit_sum, "%")
```

#### Диаграмма размаха для продаж по платформам

***Диаграммы размаха для топовых платформ*** 

```python
print(f'Лидирующая группа платформ с наибольшими суммарными продажами (TOP-{top_platform}):')
print(top_platform_list,'\n')
```

Для каждой платформы определим метрики данных, рассчитаем границу в Q3+ 1.5 IQR,  посчитаем количество и сумму продаж в датасете после  границы и в целом по платформе  и выведем диаграмму размаха с заданной границей.

```python
box_lim = 2  #  верхний предел диапазона данных для диаграммы размаха

title_base = 'Диаграмма размаха  продаж  для топовой платформы ' 

for  tp in top_platform_list:
    title_res = title_base + tp
    
    sales = data_actual[data_actual['platform']== tp] #фильтруем датасет по платформе

    # метрики, границы и продажи для платформы
    sales_describe =sales['all_sales'].describe()
    print('Метрики данных по глобальным продажам для топовой платформы',tp,'\n' , sales_describe)
    
    iqr_limit = sales_describe[6] + 1.5*(sales_describe[6] - sales_describe[4])
    print("\n Граница в Q3+ 1.5 IQR:", round(iqr_limit ,2) ,'\n')
    
    count_all = sales['all_sales'].count()
    limit_count_after = sales[sales['all_sales']> iqr_limit]['all_sales'].count()
    limit_count_ratio = round(limit_count_after/count_all ,2)
    
    sum_all = sales['all_sales'].sum()
    limit_sum_after = sales[sales['all_sales']> iqr_limit]['all_sales'].sum()
    limit_sum_ratio = round(limit_sum_after/sum_all, 2)
    
    print("Отношение количества данных после  границы (Q3+ 1.5 IQR)  ко всем продажам платформы:  ", round(100*limit_count_ratio ,2) , "%")
    print("Отношение суммы продаж после  границы (Q3+ 1.5 IQR)  ко всем продажам платформы:  ", round(100*limit_sum_ratio ,2) , "%")
 
    # диаграмма размаха
    fig, ax = plt.subplots(figsize=(17, 3))
    ax.boxplot(sales['all_sales'], vert = 0)
    plt.title(title_res)
    plt.xlabel('Объем продаж, млн. копий')
    plt.xlim(0,box_lim )
    plt.show()
    print('\n\n')
    
    
```

 Наблюдаем очень неравномерный характер продаж. Много отдельных  игр с небольшими продажами и небольшая доля игр, продажи которых очень заметны на фоне остальных.

```python
piv_table_sales_platform = data.pivot_table(index=['platform'], values=['all_sales'], aggfunc= 'sum')
piv_table_sales_platform = piv_table_sales_platform.reset_index()
piv_table_sales_platform = piv_table_sales_platform.rename(columns={'index':'platform'})

sales_platform_sorted= piv_table_sales_platform.sort_values(by='all_sales', ascending=False)
sales_platform_sorted = sales_platform_sorted.reset_index(drop=True)

# определим количество платформ, входящих в топ.
top_platform_score = 3  # это значение можно(!) менять- автоматически сформируется топ платформ.  Лучше не больше 4-5.
# Если значение изменено, то необходимо перезапустить последующие ячейки, так как они используют перечень топовых платформ.

top_platform_score_list = list(sales_platform_sorted[0:top_platform_score]['platform'])
print(f'Лидирующая группа платформ с наибольшими суммарными продажами (TOP-{top_platform_score}):')
print(top_platform_score_list)
```

```python
title_base_c = 'Уровнень продаж в зависимости от оценок критиков (без заполнения пропусков)  для платформы  ' 
title_base_u = 'Уровнень продаж в зависимости от оценок пользователей (без заполнения пропусков) для платформы  '
#title_base_c_null = 'Уровнень продаж в зависимости от оценок критиков (с заполнением пропусков) для платформы  ' 
#title_base_u_null = 'Уровнень продаж в зависимости от оценок пользователей (с заполнением пропусков) для платформы  '


for  tp in top_platform_score_list:
    title_res_c = title_base_c + tp
    title_res_u = title_base_u + tp
#    title_res_c_null = title_base_c_null + tp
#    title_res_u_null = title_base_u_null + tp
    
    piv_t_plat_score_c = data_top_score_filtered[data_top_score_filtered['platform']==tp].pivot_table(index=['critic_score'], values=['all_sales'], aggfunc= 'sum')
    piv_t_plat_score_u = data_top_score_filtered[data_top_score_filtered['platform']==tp].pivot_table(index=['user_score'], values=['all_sales'], aggfunc= 'sum')
        
#    piv_t_plat_score_c_null = data_top_score_filtered[data_top_score_filtered['platform']==tp].pivot_table(index=['critic_score_null'], values=['all_sales'], aggfunc= 'sum')
#    piv_t_plat_score_u_null = data_top_score_filtered[data_top_score_filtered['platform']==tp].pivot_table(index=['user_score_null'], values=['all_sales'], aggfunc= 'sum')
    
    piv_t_plat_score_c = piv_t_plat_score_c.reset_index()
    piv_t_plat_score_u = piv_t_plat_score_u.reset_index()
#    piv_t_plat_score_c_null = piv_t_plat_score_c_null.reset_index()
#    piv_t_plat_score_u_null = piv_t_plat_score_u_null.reset_index()
    
    piv_t_plat_score_c = piv_t_plat_score_c.rename(columns={'index':'critic_score'})
    piv_t_plat_score_u = piv_t_plat_score_u.rename(columns={'index':'user_score'})
#    piv_t_plat_score_c_null = piv_t_plat_score_c_null.rename(columns={'index':'critic_score_null'})
#    piv_t_plat_score_u_null = piv_t_plat_score_u_null.rename(columns={'index':'user_score_null'})
    
    print()
    pirson_critic_score = piv_t_plat_score_c['critic_score'].corr(piv_t_plat_score_c['all_sales'])
    print(f'Корреляция между оценкой критиков и уровнем продаж  для платформы {tp} : {round(pirson_critic_score, 2)}')
    
    piv_t_plat_score_c.plot(x='critic_score', y='all_sales', xlim= (0, 100), kind='scatter',alpha = 0.8,  figsize=(17,4) , title =title_res_c , color= 'blue')
#    piv_t_plat_score_c_null.plot(x='critic_score_null', y='all_sales', xlim= (0, 100), kind='bar', figsize=(17,2) , title =title_res_c_null , color= 'royalblue')
    plt.show()
    
    print()
    pirson_user_score = piv_t_plat_score_u['user_score'].corr(piv_t_plat_score_u['all_sales'])
    print(f'Корреляция между оценкой пользователей и уровнем продаж  для платформы {tp} : {round(pirson_user_score, 2)}')
    
    piv_t_plat_score_u.plot(x='user_score', y='all_sales', xlim= (0, 10),   kind='scatter',alpha = 0.8,  figsize=(17,4) , title =title_res_u, color = 'green' ) 
#    piv_t_plat_score_u_null.plot(x='user_score_null', y='all_sales', xlim= (0, 10),   kind='bar', figsize=(17,2) , title =title_res_u_null, color = 'darkcyan' )
    plt.show()
    
    print('\n \n')
```

## Статистический анализ данных

В рамках исследования проверяем гипотезы на выборке данных за 2015-2016 годы, предполагаем, что все игры имеют оценку (берем только данные с оценкой).

### Пользовательские рейтинги платформ Xbox One  и PC

1. В рамках исследования необходимо проверить гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  
2. В рамках исследования проводится проверка гипотезы о равенстве средних двух генеральных совокупностей, по взятым из них выборкам. Поэтому применяем T-тест ttest_ind() библиотеки scipy.

 Нулевая гипотеза H_0: пользовательские рейтинги платформ Xbox One и PC одинаковые.  
 Альтернативная гипотеза H_а: пользовательские рейтинги платформ Xbox One и PC не одинаковые.  
 Уровень значимости alpha = 0.05.
 

```python
#вызов метода для проверки гипотезы
results = st.ttest_ind(data[(data['platform'] =='XOne') & (~data['user_score'].isna()) & (data['year_of_release']>=2015) ]['user_score'],data[(data['platform'] =='PC') & (~data['user_score'].isna()) & (data['year_of_release']>=2015) ]['user_score'],equal_var = False )
 

#уровeнь значимости
alpha = 0.05

# вывод значения p-value на экран
print( results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print()    
xone_mean = round(data[(data['platform'] =='XOne') & (~data['user_score'].isna()) & (data['year_of_release']>=2015)  ]['user_score'].mean() , 2)
pc_mean =  round(data[(data['platform'] =='PC') & (~data['user_score'].isna()) & (data['year_of_release']>=2015) ]['user_score'].mean() , 2)

print('Пользовательский средний рейтинг платформы Xbox One: ', xone_mean)   
print('Пользовательский средний рейтинг платформы PC: ', pc_mean)  
```

Пользовательские рейтинги платформ Xbox One и PC одинаковые с вероятностью 29.46 %, что ощутимо выше уровня значимости 5%,   рейтинги действительно близкие к одинаковым.

### Пользовательские рейтинги жанров Action  и Sports

 В рамках исследования необходимо проверить гипотезу о том, что средние пользовательские рейтинги жанров Action и  Sports разные.

Нулевая гипотеза H_0: пользовательские рейтинги жанров Action и Sports  одинаковые.  
 Альтернативная гипотеза H_а: пользовательские рейтинги жанров Action и Sports  разные.  
 Уровень значимости alpha = 0.05.
 

```python
#вызов метода для проверки гипотезы
results = st.ttest_ind(data[(data['genre'] =='Action') & (~data['user_score'].isna()) & (data['year_of_release']>=2015)  ]['user_score'], data[(data['genre'] =='Sports') & (~data['user_score'].isna()) & (data['year_of_release']>=2015)  ]['user_score'] , equal_var = False )
 
#уровeнь значимости
alpha = 0.05

# вывод значения p-value на экран
print( results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print()    
action_mean = round(data[(data['genre'] =='Action') & (~data['user_score'].isna())  & (data['year_of_release']>=2015) ]['user_score'].mean() , 2)
sports_mean =  round(data[(data['genre'] =='Sports') & (~data['user_score'].isna()) & (data['year_of_release']>=2015)  ]['user_score'].mean() , 2)

print('Пользовательский средний рейтинг жанра Action: ', action_mean)   
print('Пользовательский средний рейтинг жанра Sports: ', sports_mean)  
```

Пользовательские рейтинги жанров Action и  Sports  одинаковые с вероятностью 0,0000000597 %, что значительно ниже уровня значимости  5%.  
Рейтинги жанров Action и  Sports различаются.

3. EC («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. 
4. RP («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.
